```python
#bai3
def tính_tam_giác(a, b, c):
    TOLERANCE = 1e-9 # Định nghĩa TOLERANCE sai số

    if not ((a + b) > c and (a + c) > b and (b + c) > a):
        print("Không phải tam giác")
        return

    # Tính toán chu vi và diện tích (Heron's formula)
    p_semi = (a + b + c) / 2  # Nửa chu vi
    s = (p_semi * (p_semi - a) * (p_semi - b) * (p_semi - c)) ** 0.5 # Diện tích
    l = a + b + c  # Chu vi

    tam_giac = "tam giác thường"
    if abs(a - b) < TOLERANCE and abs(b - c) < TOLERANCE:
        tam_giac = "tam giác đều"
    else:
        is_vuong = False
        if abs(a*a + b*b - c*c) < TOLERANCE or \
           abs(a*a + c*c - b*b) < TOLERANCE or \
           abs(b*b + c*c - a*a) < TOLERANCE:
            is_vuong = True
        is_can = False
        if abs(a - b) < TOLERANCE or abs(a - c) < TOLERANCE or abs(b - c) < TOLERANCE:
            is_can = True
        if is_vuong and is_can:
            tam_giac = "tam giác vuông cân"
        elif is_vuong:
            tam_giac = "tam giác vuông"
        elif is_can:
            tam_giac = "tam giác cân"
    print("Kết quả:")
    print("Loại tam giác:", tam_giac)
    print("Chu vi:", l)
    print("Diện tích:", s)
try:
    print("Nhập 3 số a, b, c bất kì cho các cạnh của tam giác:")
    a = float(input("Cạnh a: "))
    b = float(input("Cạnh b: "))
    c = float(input("Cạnh c: "))
    tính_tam_giác(a, b, c)
except ValueError:
    print("Lỗi: Đầu vào không phải là số hợp lệ. Vui lòng nhập các giá trị số.")
except Exception as e:
    print(f"Đã xảy ra lỗi không mong muốn: {e}")
```

```python
def giai_phuong_trinh_bac_nhat():   #bai4
    print("Chương trình giải phương trình bậc nhất (ax + b = 0)")
    try:
        a = float(input("Nhập số thực a: "))
        b = float(input("Nhập số thực b: "))
    except ValueError:
        print("Lỗi đầu vào: a và b phải là số thực.")
        return # Thoát khỏi hàm nếu đầu vào không hợp lệ

    print(f"Giải phương trình {a}x + {b} = 0")

    if a != 0:
        x = -b / a
        print(f"Nghiệm duy nhất của phương trình là: {x:.2f}")
    elif a == 0 and b == 0:
        print("Phương trình vô số nghiệm")
    elif a == 0 and b != 0:
        print("Phương trình vô nghiệm")

giai_phuong_trinh_bac_nhat()
```

```python
#bai8
def giai_he_phuong_trinh_tuyen_tinh():
  print("chương trình giải hệ phương trình tuyến tính")
  try:
    a,b,c,d,m,n=map(float,input().split())
  except ValueError:
    print("lỗi đầu vào:a,b,c,d,m,n phải là các số thực.")
    return

  if a*d-b*c == 0:
    if (a*n - m*c == 0) and (b*n - m*d == 0):
        print("phương trình vô số nghiệm")
    else:
        print("phương trình vô nghiệm")
  else:
    x=(m*d-b*n)/(a*d-b*c)
    y=(a*n-m*c)/(a*d-b*c)
    print("phương trình có nghiệm duy nhất x=",x,"và y=",y)

giai_he_phuong_trinh_tuyen_tinh()
```
